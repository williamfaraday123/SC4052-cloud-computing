## 3 sample TCP algorithms

```python
import numpy as np

def tcp_reno(packet_loss_rate, bandwidth):
    """
    Simulate TCP Reno behavior.
    :param packet_loss_rate: Probability of packet loss (0 to 1).
    :param bandwidth: Available bandwidth (Mbps).
    :return: Throughput (Mbps), Latency (ms).
    """
    cwnd = 10  # assume initial congestion window size
    throughput = bandwidth * (1 - packet_loss_rate)  # assume simplified throughput calculation
    latency = 1000 / throughput  # assume simplified latency calculation (ms)
    return throughput, latency

def dctcp(packet_loss_rate, bandwidth):
    """
    Simulate DCTCP behavior.
    :param packet_loss_rate: Probability of packet loss (0 to 1).
    :param bandwidth: Available bandwidth (Mbps).
    :return: Throughput (Mbps), Latency (ms).
    """
    # DCTCP uses ECN (Explicit Congestion Notification) for finer congestion control
    throughput = bandwidth * (1 - packet_loss_rate ** 2)  # assume more aggressive than Reno
    latency = 800 / throughput  # assume lower latency due to better congestion control
    return throughput, latency

def timely(packet_loss_rate, bandwidth):
    """
    Simulate TIMELY behavior.
    :param packet_loss_rate: Probability of packet loss (0 to 1).
    :param bandwidth: Available bandwidth (Mbps).
    :return: Throughput (Mbps), Latency (ms).
    """
    # TIMELY uses RTT-based congestion control
    throughput = bandwidth * (1 - packet_loss_rate)  # assume similar to Reno but with RTT adjustments
    latency = 500 / throughput  # assume lower latency due to RTT-based control
    return throughput, latency
```

## Test the 3 TCP algorithms

```python
# Test TCP Reno
throughput_reno, latency_reno = tcp_reno(0.1, 100)

# Test DCTCP
throughput_dctcp, latency_dctcp = dctcp(0.1, 100)

# Test TIMELY
throughput_timely, latency_timely = timely(0.1, 100)
```

## Compare results: Use matplotlib to visualize the results:

```python
import matplotlib.pyplot as plt

# Data for plotting
algorithms = ['TCP Reno', 'DCTCP', 'TIMELY']
throughputs = [throughput_reno, throughput_dctcp, throughput_timely]
latencies = [latency_reno, latency_dctcp, latency_timely]

# Plot Throughput
plt.figure()
plt.bar(algorithms, throughputs, color=['blue', 'green', 'red'])
plt.title('Throughput Comparison')
plt.ylabel('Throughput (Mbps)')
plt.show()

# Plot Latency
plt.figure()
plt.bar(algorithms, latencies, color=['blue', 'green', 'red'])
plt.title('Latency Comparison')
plt.ylabel('Latency (ms)')
plt.show()
```
